```python
#imports

import matplotlib.pyplot as plt
import numpy as np
import cv2
from skimage import data, img_as_float, io, segmentation, color
from skimage.segmentation import chan_vese
from skimage.filters import sobel
from skimage.color import rgb2gray
from skimage.future import graph
import SimpleITK as sitk
```

```python
#Reading the image and info

img = sitk.ReadImage('D:\\dataset\\xVertSeg.v1\\Data1\\images\\image001.mhd')
imgArr = sitk.GetArrayFromImage(img)
org = np.array(list(reversed(img.GetOrigin())))
spacing = np.array(list(reversed(img.GetSpacing())))
```

```python
#visualizing

'''rgbImage = sitk.Image([128, 128], sitk.sitkVectorUInt8, 3)
rgbImage

arr = sitk.GetArrayFromImage(rgbImage)
image = sitk.GetImageFromArray(arr, isVector=True)'''

view = sitk.GetArrayViewFromImage(img)[0,:,:]

plt.imshow(view, cmap="gray")
view.shape
```

```python
#Chanvese Seg
image = view

cv = chan_vese(image, mu=0.01, 
               lambda1=1, lambda2=1, tol=1e-3, 
               max_iter=100,
               dt=0.5, init_level_set="checkerboard", extended_output=True)

fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.flatten()

ax[0].imshow(image, cmap="gray")
ax[0].set_axis_off()
ax[0].set_title("Original Image", fontsize=12)

ax[1].imshow(cv[0], cmap="gray")
ax[1].set_axis_off()
title = "Chan-Vese segmentation - {} iterations".format(len(cv[2]))
ax[1].set_title(title, fontsize=12)

ax[2].imshow(cv[1], cmap="gray")
ax[2].set_axis_off()
ax[2].set_title("Final Level Set", fontsize=12)

ax[3].plot(cv[2])
ax[3].set_title("Evolution of energy over iterations", fontsize=12)

fig.tight_layout()
plt.show()
```

```python
final = np.array(cv[1])
final[final < 2.45] = 0

plt.imshow(final, cmap="gray")
```
